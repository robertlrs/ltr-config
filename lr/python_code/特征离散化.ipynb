```python
from pyspark import SparkConf, SparkContext

from pyspark.sql import SparkSession
import numpy as np
import math

spark = SparkSession.builder.appName("lr_sample").getOrCreate()
```

```python
def ctr():
    ctr = spark.sql("SELECT (click+1) / (exposure+50) as ctr from songwt.item_click_exposure3 WHERE item_id IS NOT NULL").collect()
    ctr_list = [list(c)[0] for c in ctr]
    myarray = np.asarray(ctr_list)
    t = np.percentile(myarray, [10, 20, 30, 40, 50, 60,70, 80, 90])
    print(t)
    
def catectr():
    ctr = spark.sql("select ctr from songwt.lr_catectr2").collect()
    ctr_list = [list(c)[0] for c in ctr]
    myarray = np.asarray(ctr_list)
    t = np.percentile(myarray, [20,40,60,80])
    print(t)
    
def cvr():
    ctr = spark.sql("SELECT pv_buy / pv_click as cvr from songwt.cvr WHERE pv_buy > 0").collect()
    ctr_list = [list(c)[0] for c in ctr]
    myarray = np.asarray(ctr_list)
    t = np.percentile(myarray, [20,40,60,80])
    print(t)

def collect():
    collect = spark.sql("select collect from (select DISTINCT iid, collect_onemonth as collect, ds from songwt.lr_sample where iid IS NOT NULL) t where collect > 0").collect()
    ctr_list = [row.collect for row in collect]
    myarray = np.asarray(ctr_list)
    t = np.percentile(myarray, [20, 40, 60, 80])
    print(t)
    
def addcart():
    addcart = spark.sql("select addcart from (select DISTINCT iid, addcart_onemonth as addcart, ds from songwt.lr_sample where iid IS NOT NULL) t where addcart > 0").collect()
    addcart_list = [row.addcart for row in addcart]
    myarray = np.asarray(addcart_list)
    t = np.percentile(myarray, [25, 50, 75])
    print(t)
    
    
def sale():
    sale = spark.sql("select sale from (select DISTINCT iid, sale, ds from songwt.sample_sale) t where sale >0  ").collect()
    sale_list = [row.sale for row in sale]
    myarray = np.asarray(sale_list)
    t = np.percentile(myarray, [25, 50, 75])
    print(t)
    


    
def refund():
    collect = spark.sql("SELECT interval_rate from zhouyf.crafts_refund where ds BETWEEN '2018-07-01' and '2018-09-30' and interval_rate <> 0 ").collect()
    ctr_list = [list(c)[0] for c in collect]
    myarray = np.asarray(ctr_list)
    t = np.percentile(myarray, [25, 50, 75])
    print(t)
    
def tag_price(tag_id):
    
    sql = "select unit_price from songwt.tag_price where tag_id=%d and unit_price > 10 " % tag_id
    print(sql)
    tag = spark.sql(sql).collect()
    price = [row.unit_price for row in tag]
    myarray = np.asarray(price)
    t = np.percentile(myarray, [20,40,60,80])
    return t

def ctr2():
    ctr = spark.sql("SELECT ctr, (click+5) / (exposure+50) as ctr2, (click+10) / (exposure+100) as ctr3 from songwt.index_ctr where click > 0").collect()
    
    ctrlist = [(row.ctr, row.ctr2, row.ctr3) for row in ctr]
    
    _result = 0
    _result2 = 0
    for elem in ctrlist:
        t = elem[1]-elem[0]
        t2 = elem[2]-elem[0]
        _result += pow(t,2)
        _result2 += pow(t2,2)
    print(_result, _result2)
    
def ctr3():
     
    ctr = spark.sql("SELECT (click+5) / (exposure+50) as ctr from songwt.index_ctr where item_id is not null").collect()

    ctrlist = [row.ctr for row in ctr]


    myarray = np.asarray(ctrlist)
    t = np.percentile(myarray, [10, 20, 30, 40, 50, 60,70, 80, 90])
    print(t)
    
def getcity():
    
    special_city = ["西双版纳", "宜兴", "景德镇"]
    
    city = spark.sql("SELECT DISTINCT city from user.craftsman where length(city) = 2").collect()
    
    city_list = [row.city for row in city]
    
    city_list.append("西双版纳")
    
    craftsman = spark.sql("select customer_id, city from user.craftsman where city is not null").collect()
    
    craftsman_list = [(row.customer_id, row.city) for row in craftsman]
    
    array = list()
    for elem in craftsman_list:
        cid = elem[0]
        city = elem[1]
        craftsman_city = ""
        if city.find("宜兴") != -1:
            craftsman_city = "宜兴"
        elif city.find("景德镇") != -1:
            craftsman_city = "景德镇"
        else:
            for item in city_list:
                if city.find(item) != -1:
                    craftsman_city = item
                    break
        if len(craftsman_city) == 0:
            craftsman_city = city
        if len(city) > 3:
            print(city, craftsman_city)
                  
        array.append((cid, city, craftsman_city))
    #print(array)
            
    
    
if __name__ == "__main__":
    #getcity()
    #collect()
    #addcart()
    sale()
  
        
        
    
```
